Como estamos fazendo uma análise de série temporal, precisamos que a coluna 'date' esteja em datetime. Portanto agora vamos fazer a conversão dessa coluna de object para datetime:

---

As we are conducting a time series analysis, we need the 'date' column to be in datetime format. Therefore, let's now convert this column from object to datetime:

```python
# Verificando o tipo dos dados das colunas ('spx','dax','ftse','nikkei') / Checking the data types of the columns ('spx', 'dax', 'ftse', 'nikkei')

print('Conferindo os tipos de dados das outras colunas:\n',df[['spx','dax','ftse','nikkei']].dtypes)
```

```python
# Transformando a coluna 'date' de object para datetime / Converting the 'date' column from object to datetime

df['date'] = pd.to_datetime(df['date'])
```

**Explicando o erro / Explaining the error:**


O erro indica que os dados na coluna 'date' não correspondem ao formato padrão ("%m/%d/%Y") esperado pela função pd.to_datetime(). Isso ocorre quando os dados da data estão em um formato diferente do esperado pelo pandas.

Neste caso, os dados estão no formato "DD/MM/AAAA" (dia/mês/ano), enquanto o pandas está esperando o formato "MM/DD/AAAA" (mês/dia/ano).

---

The error indicates that the data in the 'date' column does not match the standard format ("%m/%d/%Y") expected by the pd.to_datetime() function. This occurs when the date data is in a different format than what pandas expects.

In this case, the data is in the "DD/MM/YYYY" (day/month/year) format, while pandas is expecting the "MM/DD/YYYY" (month/day/year) format.

```python
# Averiguado os dados da coluna da coluna date / Averiguado os dados da coluna da coluna date

df.date
```

*Corrigindo o erro / Fixing the error:*

```python
#Transformando a coluna 'date' de object para datetime, com a função formatada para dia/mês/ano / Converting the 'date' column from object to datetime, with the function formatted for day/month/year

df['date'] = pd.to_datetime(df['date'],format='%d/%m/%Y')
```

```python
# Certificando se a conversão do tipo de dados da coluna 'date' foi realizada com sucesso / Ensuring the data type conversion of the 'date' column was successful

print('Inspecionando o tipo de dados: ',df['date'].dtypes)
```

Com a coluna 'date' no formato de datetime, podemos começar a nossa análise de Série Temporal.

```python
## Criação do gráfico / Creating the chart

# Configurando o estilo dos gráficos / Configuring the chart style

sns.set(style='whitegrid')

# Plotando os gráficos de linha para visualizar a tendência dos índice / Plotting line charts to visualize index trends

plt.figure(figsize=(14,7)) # Configurando o tamanho do gráfico / Setting the chart size

plt.plot(df['date'],df['spx'],label='S&P 500') # Plotando a linha da coluna 'spx' / Plotting the line for the 'spx' column
plt.plot(df['date'],df['dax'],label='DAX') # Plotando a linha da coluna 'dax' / Plotting the line for the 'dax' column
plt.plot(df['date'],df['ftse'],label='FTSE') # Plotando a linha da coluna 'ftse' / Plotting the line for the 'ftse' column
plt.plot(df['date'],df['nikkei'],label='NIKKEI') # Plotando a linha da coluna 'nikkei' / Plotting the line for the 'nikkei' column

plt.xlabel('Data') # Nomeando o eixo X / Naming the X-axis
plt.ylabel('Índice de Preço') # Nomeando o eixo Y /Naming the Y-axis
plt.title('Tendências dos Índices ao longo do tempo',loc='left',color='gray',fontsize= 18) # Colocando o título /Adding a title
plt.legend() # Inserindo legenda no gráfico / Inserting a legend on the chart
plt.show()
```

```python
# Criando uma função para quatro gráficos / Creating a function for four plots

def plot_comparacao(x, y1, y2, y3, y4,
                    subtitulo1, subtitulo2, subtitulo3, subtitulo4,
                    dataset, titulo, labelx, labely):
    '''
    Função para criar uma figura com quatro gráficos lineares em um único layout.

    Parâmetros:
    • x -> Mantém essa variável para todos os gráficos, pois será a mesma para todos;
    • y1, y2, y3, y4 -> Variáveis distintas, pois cada gráfico terá uma coluna do DataFrame;
    • labelx -> Variável compartilhada do eixo x para todos os gráficos.
    • labely -> O eixo y será o mesmo para todos os gráficos;
    • subtitulo1, subtitulo2, subtitulo3, subtitulo4 -> Títulos para cada gráfico.
    • dataset -> DataFrame utilizado para plotar os gráficos.
    • titulo -> Título geral para os quatro gráficos.

    ---

    Function to create a figure with four line plots in a single layout.

    Parameters:

    • x: Keep this variable for all plots, as it will be the same for all.
    • y1, y2, y3, y4: Distinct variables, as each plot will have a column from the DataFrame.
    • labelx: Shared x-axis variable for all plots.
    • labely: The y-axis will be the same for all plots.
    • subtitulo1, subtitulo2, subtitulo3, subtitulo4: Titles for each plot.
    • dataset: DataFrame used to plot the graphs.
    • titulo: General title for the four plots.


    '''

    # Espaço, estilo e título para os 4 gráficos / Space, style, and title for the 4 plots
    plt.figure(figsize=(16,12))  # Tamanho do espaço e dos gráficos / Size of space and plots
    sns.set(style='whitegrid')   # Estilo do fundo dos gráficos / Style of the background of the plots
    plt.suptitle(titulo, fontsize=18, color='black')  # Título geral dos gráficos / General title of the plots

    # 1° Gráfico / 1st Plot
    ax1 = plt.subplot(4,1,1)  # Posição do gráfico / Plot position
    sns.lineplot(x=x, y=y1, data=dataset, ax=ax1)  # Plotando o gráfico, ax=ax1 é para não haver confusão na hora da plotagem / Plotting the graph, ax=ax1 is to avoid confusion during plotting
    ax1.set_title(subtitulo1, fontsize=14, loc='center', color='gray')  # Configurando o subtitulo / Configuring the subtitle
    ax1.set_xlabel(labelx, fontsize=10)  # Configurando o eixo x / Configuring the x-axis
    ax1.set_ylabel(labely, fontsize=10)  # Configurando o eixo y / Configuring the y-axis

    # 2° Gráfico / 2nd Plot
    ax2 = plt.subplot(4,1,2)
    sns.lineplot(x=x, y=y2, data=dataset, ax=ax2)
    ax2.set_title(subtitulo2, fontsize=14, loc='center', color='gray')
    ax2.set_xlabel(labelx, fontsize=10)
    ax2.set_ylabel(labely, fontsize=10)

    # 3° Gráfico / 3rd Plot
    ax3 = plt.subplot(4,1,3)
    sns.lineplot(x=x, y=y3, data=dataset, ax=ax3)
    ax3.set_title(subtitulo3, fontsize=14, loc='center', color='gray')
    ax3.set_xlabel(labelx, fontsize=10)
    ax3.set_ylabel(labely, fontsize=10)

    # 4° Gráfico / 4th Plot
    ax4 = plt.subplot(4,1,4)
    sns.lineplot(x=x, y=y4, data=dataset, ax=ax4)
    ax4.set_title(subtitulo4, fontsize=14, loc='center', color='gray')
    ax4.set_xlabel(labelx, fontsize=10)
    ax4.set_ylabel(labely, fontsize=10)

    # Ajustar layout e mostrar os gráficos / Adjust layout and show the plots
    plt.tight_layout()  # Função usada para ajustar automaticamente o espaçamento entre os subplots / Function used to automatically adjust the spacing between subplots
    plt.show()  # Exibir o gráfico / Display the plot
```

**Analisando os gráficos:**

Podemos observar como as crises financeiras globais impactaram esses mercados, apesar de sua robustez. Todos experimentaram quedas abruptas por volta dos anos de 2004 e 2008, com destaque para o ano de 2008, quando esses quatro importantes mercados sofreram suas maiores quedas.
